```python
import cv2
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("/content/bina.jpg")
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
parlak_img= cv2.convertScaleAbs(img, beta=80) #beta=80 parlaklik parametresidir.
```

```python
plt.figure(figsize=(15,20))
plt.subplot(121), plt.imshow(img), plt.title("orijinal")
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(parlak_img), plt.title("parlak")
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
cv2.imwrite("parlak bina.jpg", parlak_img)  #yeni dosya olusturup parlak resmi kaydettik.
```

```python
yaprak_img=cv2.imread("/content/yaprak.jpg")
yaprak_img=cv2.cvtColor(yaprak_img, cv2.COLOR_BGR2RGB)
```

```python
kontrast_img=cv2.convertScaleAbs(yaprak_img, alpha=1.4) #deger arttigi icin renk daha faza gozukeceks
```

```python
plt.figure(figsize=(15,20))
plt.subplot(121), plt.imshow(yaprak_img), plt.title("orijinal")
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(kontrast_img), plt.title("kontrast")
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
parlak_gray = cv2.convertScaleAbs(gray_img, beta = 100)
plt.imshow(parlak_gray, cmap='gray')
```

histogram ile renkleri orantili bir sekilde dagitalim. histogram esitleme gri tonlamali fotograflarda kullanilir. renkli goruntulerde calismaz.

```python
hist_gray = cv2.equalizeHist(gray_img)
```

```python
plt.figure(figsize=(15,20))

plt.subplot(121), plt.imshow(parlak_gray,cmap='gray'), plt.title('Parlak Gri')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(hist_gray, cmap='gray'), plt.title('Histogram esitleme')
plt.xticks([]), plt.yticks([])

plt.show()
```
